# Step 6: Data Visualization

```python
plt.figure(figsize=(25,8))
plt.bar('Sub-Category','Category', data=sample1)
plt.title('Category vs Sub Category')
plt.xlabel('Sub-Catgory')
plt.ylabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.bar('Category','Sales', data=sample1)
plt.title('Category vs Sales')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

**From This chart we can inference that Technology has more sales over any other category. So there is need of good marketing stretegy in the Furniture department**

```python
plt.figure(figsize=(16,8))
plt.bar('Category','Profit', data=sample1)
plt.title('Category vs Profit')
plt.xlabel('Category')
plt.ylabel('profit')
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Sales', data=sample1)
plt.title('Sub-Category vs Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.bar('Region','Sales', data=sample1)
plt.title('Region vs Sales')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

**Sales in the East region are low compare to any other region**

```python
plt.figure(figsize=(16,8))
plt.bar('Region','Profit', data=sample1)
plt.title('Region vs Profit')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.show()
```

# Segment wise analysis

```python
ps=sample.groupby('Segment')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['green','red'],figsize=(25,12))
plt.title('Profit/Loss and Sales across Segment')
plt.xlabel('Segment',fontsize=15)
plt.ylabel('Profit/Loss and Sales',fontsize=15)
plt.show()
```

# State Wise

```python
plt.figure(figsize=(15,12))
sns.countplot(x='State',data=sample,palette='rocket_r',order=sample['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

**Here we can see that in california highest number of orders are received. It is the most active state for selling.**

# Profit and Discount relation

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=sample)
plt.legend()
plt.show()
```

**We can see from line graph that at .5 discount profit is minimum**

# Profit and Quantity Relation

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=sample)
plt.legend()
plt.show()
```

**Higher the quantity higher the profit**

# Segment Wise Sale in each Region

```python
plt.figure(figsize=(12,8))
plt.title('Segment Wise Sales in Each Region')
sns.barplot(x='Region',y='Sales',data=sample,hue='Segment',order=sample['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region',fontsize=15)
plt.ylabel('Sales',fontsize=15)
plt.show()
```

# P/L and Sales by States

```python
ps=sample.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['green','red'],figsize=(25,12))
plt.title('Profit/Loss and Sales across States')
plt.xlabel('States',fontsize=15)
plt.ylabel('Profit/Loss and Sales',fontsize=15)
plt.show()
```

# P/L and Sales by Sub-Category

```python
ps=sample.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sub-Category',ascending=False)
ps[:].plot.bar(color=['green','red'],figsize=(25,12))
plt.title('Profit/Loss and Sales across Sub-Category')
plt.xlabel('Sub-Category',fontsize=25)
plt.ylabel('Profit/Loss and Sales',fontsize=15)
plt.show()
```
